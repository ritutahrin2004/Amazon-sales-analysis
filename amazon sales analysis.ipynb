```python
# Convert all values to string, remove '%', then convert to float
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)

# Handle non-convertible values safely (e.g., 'nan', etc.)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
```

```python
print("Correlation with Rating:")
print(df['discount_percentage'].corr(df['rating']))

print("\nCorrelation with review count:")
print(df['discount_percentage'].corr(df.groupby('product_id')['review_id'].transform('count')))
```

```python
print(df['discount_percentage'].describe())
```

```python
# Step 1: Check for non-null values
print(df['discount_percentage'].isna().sum(), df['rating'].isna().sum())

# Step 2: Drop rows with missing values in either column
df_corr = df[['discount_percentage', 'rating']].dropna()

# Step 3: Check again
print("Correlation with Rating:", df_corr['discount_percentage'].corr(df_corr['rating']))

# Step 4: Review count correlation
df['review_count'] = df.groupby('product_id')['review_id'].transform('count')
df_corr2 = df[['discount_percentage', 'review_count']].dropna()
print("Correlation with review count:", df_corr2['discount_percentage'].corr(df_corr2['review_count']))
```

```python
print("Unique discount percentages:", df['discount_percentage'].nunique())
print("Sample values:\n", df['discount_percentage'].dropna().unique()[:10])
```

```python
# Strip whitespace, remove %, replace empty strings or bad entries with NaN
import numpy as np
df['discount_percentage'] = (
    df['discount_percentage']
    .astype(str)
    .str.strip()
    .str.replace('%', '', regex=False)
    .replace('', np.nan)
)

# Convert to numeric
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Check again
print("Unique discount percentages:", df['discount_percentage'].nunique())
print(df['discount_percentage'].dropna().value_counts().head())
```

```python
# Sort by price for a meaningful line plot
df_sorted = df[['discounted_price', 'rating']].dropna().sort_values('discounted_price')

plt.figure(figsize=(10, 5))
sns.lineplot(x='discounted_price', y='rating', data=df_sorted)
plt.title("Price vs Rating")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Clean up
df_bubble = df[['review_id', 'rating', 'discounted_price']].dropna()
df_bubble['review_count'] = df.groupby('product_id')['review_id'].transform('count')

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_bubble,
    x='review_count',
    y='rating',
    size='discounted_price',
    hue='discounted_price',
    sizes=(20, 400),
    palette='viridis',
    alpha=0.6,
    edgecolor='gray'
)
plt.title("Bubble Chart: Rating vs Review Count (Bubble Size = Price)")
plt.xlabel("Review Count")
plt.ylabel("Rating")
plt.legend(title='Price', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
